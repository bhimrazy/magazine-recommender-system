## Magazine Subscriptions
We have used Magazine subscriptions dataset from amazon review data for our project.

Ratings & reviews: https://jmcauley.ucsd.edu/data/amazon_v2/categoryFiles/Magazine_Subscriptions.json.gz

Magazines metadata: https://jmcauley.ucsd.edu/data/amazon_v2/metaFiles2/meta_Magazine_Subscriptions.json.gz


```python
# importing libraries

import pandas as pd

# ratings_data_path = "https://jmcauley.ucsd.edu/data/amazon_v2/categoryFiles/Magazine_Subscriptions.json.gz"
# magazines_data_path = "https://jmcauley.ucsd.edu/data/amazon_v2/metaFiles2/meta_Magazine_Subscriptions.json.gz"

ratings_data_path = "data/Magazine_Subscriptions.json.gz"
magazines_data_path = "data/meta_Magazine_Subscriptions.json.gz"

ratings_df = pd.read_json(ratings_data_path, lines=True)
magazines_df = pd.read_json(magazines_data_path, lines=True)
```

```python
def get_memory_usage(df:pd.DataFrame, display:bool=True) -> float:
    # Calculate memory usage of each column in bytes
    memory_usage = df.memory_usage(deep=True)
    # Sum memory usage of all columns and convert to MB
    total_memory_usage = memory_usage.sum() / (1024*1024)
    if display:
        print(f'Total memory usage: {total_memory_usage:.2f} MB')
```

```python
# selecting columns
selected_columns = ['reviewerID','asin','overall']
r_df = ratings_df[selected_columns]

# renaming columns
r_df = r_df.rename(columns={
    "reviewerID":"user_id",
    "asin":"magazine_id",
    "overall":"ratings"
})


get_memory_usage(r_df)

r_df
```

```python
# changing types of the columns
r_df = r_df.drop_duplicates()
r_df
```

```python
r_df.info()
```

```python
# changing types of the columns
column_types={
    "user_id":"string",
    "magazine_id":"category",
    "ratings":"int8",
}
new_rdf = r_df.astype(column_types)
get_memory_usage(new_rdf)
new_rdf.info()
```

```python
# write the DataFrame to a CSV file
new_rdf.to_csv('magazines_rating.csv', index=False)
```

```python
# finding popular magazines
pdf = new_rdf.groupby('magazine_id').count()["ratings"].reset_index()
prdf = new_rdf[["magazine_id","ratings"]].groupby('magazine_id').mean()['ratings'].reset_index()

new_pdf = pdf.merge(prdf, on="magazine_id")
popular_magazines = (new_pdf[new_pdf["ratings_x"]>100]
.rename(columns={'ratings_x': 'num_ratings','ratings_y': 'avg_ratings'})
.sort_values(ascending=False,by="num_ratings")
.reset_index()
.drop(columns=['index'],axis=1)
.drop_duplicates()
)
```

```python
# finding magazines with atleast 5 reviews
(
    magazines_df
    [['asin', 'description', 'brand','details', 'main_cat', 'imageURL', 'imageURLHighRes']]
    .drop_duplicates(subset=['asin']).reset_index()  
    .groupby("asin")

    # .head()
)
```

```python
popular_df = (popular_magazines
.merge(magazines_df,left_on="magazine_id",right_on="asin",how="inner")
.drop_duplicates(subset=['magazine_id']).reset_index()
[['magazine_id', 'num_ratings', 'avg_ratings', 'description', 'brand','details', 'main_cat', 'imageURL', 'imageURLHighRes']]
)
popular_df = (
popular_df[popular_df['imageURL'].str.len()>0]
.reset_index()
.drop(columns=['index'],axis=1)
)

popular_df
```

```python
import pickle
pickle.dump(popular_df, open('popular_magazines.pkl', 'wb'))
```

```python
magazines_with_at_least_5_ratings = (new_pdf[new_pdf["ratings_x"]>5]
.rename(columns={'ratings_x': 'num_ratings','ratings_y': 'avg_ratings'})
.sort_values(ascending=False,by="num_ratings")
.reset_index()
.drop(columns=['index'],axis=1)
.drop_duplicates()
)
```

```python
m_df = (magazines_with_at_least_5_ratings
.merge(magazines_df,left_on="magazine_id",right_on="asin",how="inner")
.drop_duplicates(subset=['magazine_id']).reset_index()
[['magazine_id', 'num_ratings', 'avg_ratings', 'description', 'brand','details', 'main_cat', 'imageURL', 'imageURLHighRes']]
)
m_df = (
m_df[m_df['imageURL'].str.len()>0]
.reset_index()
.drop(columns=['index'],axis=1)
)

m_df
```

```python
import pickle
pickle.dump(m_df, open('magazines.pkl', 'wb'))
```
