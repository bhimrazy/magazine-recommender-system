```python
from surprise import Dataset, Reader
from surprise import KNNBasic
from surprise.model_selection import train_test_split
from surprise import accuracy
import pandas as pd

# Load the data from a pandas dataframe
data_path = 'https://raw.githubusercontent.com/bhimrazy/recSys-project-fuseai/main/notebooks/magazines_rating.csv?token=GHSAT0AAAAAABZ5SS2QDP4PBPCVASRXD5UOZCTVLSA'
data = pd.read_csv(data_path)

# Define the rating scale
reader = Reader(rating_scale=(1, 5))

# Load the data into the surprise dataset format
dataset = Dataset.load_from_df(data[['user_id', 'magazine_id', 'ratings']], reader)

# Split the dataset into a training set and a test set
trainset, testset = train_test_split(dataset, test_size=0.2)

# Define the KNN algorithm with cosine similarity
sim_options = {'name': 'cosine', 'user_based': False}
algo = KNNBasic(sim_options=sim_options)

# Train the algorithm on the training set
algo.fit(trainset)

# Make predictions on the test set
predictions = algo.test(testset)

# Calculate the root mean squared error (RMSE)
rmse = accuracy.rmse(predictions)
print("RMSE:", rmse)

# Calculate the mean average precision (MAP)
map_ = accuracy.fcp(predictions)
print("MAP:", map_)

```
